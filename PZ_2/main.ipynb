```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('dataset.csv')
data['LastUpdated'] = pd.to_datetime(data['LastUpdated'])
data['LastUpdated'] = data['LastUpdated'].dt.date
data.drop_duplicates(subset='LastUpdated', keep='first', inplace=True)
data.reset_index(drop=True, inplace=True)
data
```

```python
import matplotlib.pyplot as plt

plt.scatter(data['LastUpdated'], data['Occupancy'])
plt.xlabel('Дата')
plt.ylabel('Количество занятых мест')
plt.title('Визуализация данных')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Calculate the regression line equation using the least square method
x = np.arange(len(data['LastUpdated']))
y = data['Occupancy']

A = np.vstack([x, np.ones(len(x))]).T
m, c = np.linalg.lstsq(A, y, rcond=None)[0]
print(m, c)

# Plot the scatter plot
plt.scatter(x, y, label='Реальные данные')

# Plot the regression line
plt.plot(x, m*x + c, 'r', label='Линейная регрессия')

# Add labels and title
plt.xlabel('День')
plt.ylabel('Количество парковочных мест')
plt.title('Метод наименьших квадратов')

# Show the plot
plt.legend()
plt.show()
```
